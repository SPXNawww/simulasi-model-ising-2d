## Tujuan

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Parameter simulasi
N = 20
T = 1.0                 #Kasus 1
jumlah_langkah = 200000
interval_pengukuran = 100

```

```python
#Menginisialisasi kisi $ N \times N $(dengan N = 20) menggunakan konfigurasi acak (Hot Start) untuk merepresentasikan entropi maksimum sistem.
# Hot Start
kisi_spin = np.random.choice(
    [-1, 1],
    size=(N, N)
)

plt.figure(figsize=(5, 5))
plt.imshow(
    kisi_spin,
    cmap='binary',
    vmin=-1,
    vmax=1
)
plt.title('Konfigurasi Awal (Hot Start)')
plt.axis('off')
plt.show()
```

```python
def langkah_metropolis(kisi_spin, T):

    # Memilih satu spin secara acak
    i = random.randint(0, N - 1)
    j = random.randint(0, N - 1)

    # Jumlah empat tetangga terdekat
    jumlah_tetangga = (
        kisi_spin[(i + 1) % N, j]
        + kisi_spin[(i - 1) % N, j]
        + kisi_spin[i, (j + 1) % N]
        + kisi_spin[i, (j - 1) % N]
    )

    # Perubahan energi lokal
    delta_E = (
        2
        * kisi_spin[i, j]
        * jumlah_tetangga
    )

    # Kriteria Metropolis
    if (
        delta_E <= 0
        or random.random() < np.exp(-delta_E / T)
    ):
        kisi_spin[i, j] *= -1

    return kisi_spin
```

```python
#Menjalankan minimal 200.000 langkah Monte Carlo untuk memastikan sistem telah mencapai kesetimbangan termal sebelum pengumpulan data makrostate dilakukan.
riwayat_M = []

for langkah in range(jumlah_langkah):

    kisi_spin = langkah_metropolis(kisi_spin, T)

    if langkah % interval_pengukuran == 0:
        M = np.mean(kisi_spin)
        riwayat_M.append(M)
```

```python

#Menampilkan visualisasi keadaan akhir kisi spin menggunakan peta warna biner (cmap='binary') berdampingan dengan kurva riwayat magnetisasi rata-rata sepanjang fungsi waktu simulasi.
fig, grafik = plt.subplots(
    1,
    2,
    figsize=(12, 5)
)

# Konfigurasi akhir spin

grafik[0].imshow(
    kisi_spin,
    cmap='binary',
    vmin=-1,
    vmax=1
)
grafik[0].set_title(
    f'Konfigurasi Akhir Spin (T = {T})'
)
grafik[0].axis('off')

# Riwayat magnetisasi
grafik[1].plot(riwayat_M)
grafik[1].set_title('Riwayat Magnetisasi')
grafik[1].set_xlabel(
    f'Langkah Monte Carlo (setiap {interval_pengukuran} langkah)'
)
grafik[1].set_ylabel(
    'Magnetisasi Rata-rata (M)'
)
grafik[1].set_ylim(-1.1, 1.1)
print(np.mean(kisi_spin))

plt.tight_layout()
plt.show()
```

```python
# Analisis Hasil

Simulasi dilakukan pada temperatur **T = 1,0**, yaitu berada di bawah temperatur kritis Model Ising dua dimensi (**Tc ≈ 2,27**). Pada kondisi ini energi interaksi antar-spin lebih dominan dibandingkan energi termal sehingga sistem cenderung meminimalkan energi total dengan menyusun spin-spin tetangga ke arah yang sama.

Konfigurasi akhir memperlihatkan terbentuknya domain magnetik yang besar dengan orientasi spin yang hampir seragam. Hal ini menunjukkan bahwa sistem berhasil mencapai keadaan teratur (ordered state). Nilai magnetisasi rata-rata berada mendekati **±1**, yang menandakan hampir seluruh spin memiliki orientasi yang sama.

Grafik riwayat magnetisasi menunjukkan bahwa pada awal simulasi magnetisasi masih berfluktuasi akibat kondisi awal Hot Start. Setelah sejumlah langkah Monte Carlo, magnetisasi menjadi relatif konstan sehingga dapat disimpulkan bahwa sistem telah mencapai kesetimbangan termal.

Hasil simulasi sesuai dengan teori Model Ising bahwa pada temperatur di bawah suhu kritis sistem berada pada **fase feromagnetik** dengan magnetisasi spontan dan keteraturan jarak jauh (*long-range order*).
```
